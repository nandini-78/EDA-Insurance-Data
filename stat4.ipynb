Calculate the mean charges for smokers who are also obese (BMI greater than or equal to 30) and have at least 2 children

```python
df[(df.smoker=='yes')&(df.bmi>=30)&(df.children>=2)]['charges'].mean()
```

Calculate the standard deviation of charges for non-smokers with a BMI between 25 and 30 who have no children.

```python
df[(df.smoker=='no')&((df.bmi>=25)|(df.bmi<=30))&(df.children==0)]['charges'].std()
```

Perform a t-test to determine if there is a significant difference in charges between smokers and non-smokers who are also obese (BMI greater than or equal to 30)

```python
import scipy.stats as st
```

```python
smoke=df[(df.smoker=='yes')&(df.bmi>=30)]['charges']
```

```python
nsmoke=df[(df.smoker=='no')&(df.bmi>=30)]['charges']
```

```python
stat,p_=st.ttest_ind(smoke,nsmoke)
```

```python
p_
```

```python
if p_>0.05:
  print("similar")
else:
  print("non similar")
```

Perform a chi-squared test to determine if there is a significant relationship between smoker status and region, but only for individuals with a BMI greater than or equal to 25.

```python
sdf=df[df['bmi']>25]
```

```python
sd=sdf[['smoker','region']]
```

```python
tab=pd.crosstab(sd.region,sd.smoker)
tab
```

```python
a,b,c,d=st.chi2_contingency(tab)                                                          
```

```python
b
```

```python
if b>0.05:
  print("no significant")
else:
  print('significant ')
```
